#### 19.HOW TO FIND THE NUMBER OF UNIVERSITIES OF ANY COUNTRY WITHOUT DUPLICATES?

```python
cross_tab2=pd.crosstab(index=data.university_name, columns=data.country)
cross_tab2
```

```python
universityofthiscountry=cross_tab2[cross_tab2['Argentina']!=0]
universityofthiscountry
```

```python
universityofthiscountry=cross_tab2[cross_tab2['India']!=0]
universityofthiscountry
```
